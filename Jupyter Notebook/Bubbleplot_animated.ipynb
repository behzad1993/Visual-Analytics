```python
#!pip install bubbly
from bubbly.bubbly import bubbleplot 
from plotly.offline import iplot

import pandas as pd
```

```python
gapminder_indicators = pd.read_csv("gapminder.tsv",sep='\t')
gapminder_indicators

figure = bubbleplot(dataset=gapminder_indicators, x_column='gdpPercap', y_column='lifeExp', 
    bubble_column='country', time_column='year', size_column='pop', color_column='continent', 
    x_title="GDP per Capita", y_title="Life Expectancy", title='Gapminder Global Indicators',
    x_logscale=True, scale_bubble=3, height=650)

iplot(figure)
```

```python
df = pd.read_pickle("event_filtered.pickle")
tmp = df[['event_id', 'event_year', 'country', 'meanPre', 'size', 'area']].groupby(['event_id', 'event_year', 'country']).sum(['meanPre', 'size', 'area']).reset_index()
tmp = tmp[['event_year', 'country', 'meanPre', 'size', 'area']].groupby(['event_year', 'country']).sum(['meanPre', 'size', 'area']).reset_index()
tmp['Continent'] = 'Other'
tmp.loc[tmp['country'] == 'IT', 'Continent'] = 'Italy'
tmp.loc[tmp['country'] == 'DE', 'Continent'] = 'Germany'
tmp.loc[tmp['country'] == 'INT', 'Continent'] = 'Mediterranean Sea'
tmp.loc[tmp['country'] == 'PL', 'Continent'] = 'Poland'
#tmp.loc[tmp['country'] == 'CZ', 'Continent'] = 'Czech Republic'
tmp.loc[tmp['country'] == 'TN', 'Continent'] = 'Tunesia'
#tmp.loc[tmp['country'] == 'AT', 'Continent'] = 'Austria'
tmp = tmp[tmp['country'] != 'INT']
tmp
```

```python
figure = bubbleplot(dataset=tmp, x_column='area', y_column='meanPre', 
    bubble_column='country', time_column='event_year', size_column='size', color_column='Continent', 
    x_title="Total Area", y_title="Total Precipitation", title='Heavy Rain Events in selected Countries',
                   #x_logscale=True, y_logscale=True
                   )

iplot(figure)
```
